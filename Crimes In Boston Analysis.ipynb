### What Types of Crimes are Most Common?

Defining a function to fetch the ten most common values in a column:

```python
def ten_frequent(column_name):
    x = crimes[column_name].value_counts().sort_values(ascending = False).head(10)
    return x
```

The <b>codes</b> dataset is a list of codes followed by the decription of the crime.

```python
codes.head()
```

It appears that the NAME column in the codes dataset is identical to the OFFENSE_DESCRIPTION column in the crimes dataset. The columns offer a brief description of the crime, while the OFFENSE_CODE_GROUP column in the crimes data tells of the type of crime.

```python
plt.figure(figsize = (10,6))
sns.set(style = 'white', context = 'talk')
most_common_desc = ten_frequent('OFFENSE_DESCRIPTION')
most_common_desc.plot.bar()
plt.xlabel('Offense Description')
plt.ylabel('Frequency')
plt.title('Ten Most Frequent Offense Descriptions');
```

It would be worth comparing the most common crime codes in the crime dataset with the name in the codes dataset.

```python
most_common_codes = ten_frequent('OFFENSE_CODE')
plt.figure(figsize = (10,6))
most_common_codes.plot.bar()
plt.xlabel('Offense Code')
plt.ylabel('Frequency')
plt.title('Ten Most Frequent Offense Codes');
```

```python
codes = codes.loc[~codes.index.duplicated(keep = 'first')]    ##Get rid of duplicate rows(indexed by CODE)
codes.loc[[3006, 3115, 3831, 1402, 802, 3301, 3410, 3114, 617, 2647]]
```

The info from the crimes dataset matches the data from the codes dataset.

```python
plt.figure(figsize = (10,6))
most_common = ten_frequent('OFFENSE_CODE_GROUP')
most_common.plot.bar()
plt.xlabel('Offense Code Group')
plt.ylabel('Frequency')
plt.title('Ten Most Frequent Offense Code Groups');
```

From the above visual, it can be seen that Motor Vehicle Accident Response is the most frequent offense code group(crime), with Larceny second at about 10,000 fewer occurrences.

### Where Are Different Types of Crimes Most Likely to Occur?

The ten most frequent types of crime are presented above, now it's just a matter of finding the most common street for each of the top ten crimes. Where are the police most likely to get called for motor vehicle accidents? Where are you most likely to get 'simply' assualted? 

```python
streets = ten_frequent('STREET')
streets
```

```python
y = crimes[crimes['STREET'] == 'WASHINGTON ST']['OFFENSE_CODE_GROUP'].value_counts().sort_values(ascending=False).head(1)
y
```

The code above shows that on Washington St. the police were called the most for larceny, with 1933 reports.

```python
y = crimes[crimes['OFFENSE_CODE_GROUP'] == 'Larceny']['STREET'].value_counts().sort_values(ascending=False).head(1)
y
```

The cell above shows that if you wanted to experience larceny in Boston, your best bet would be to go to Boylston St. since there were 2163 instances of such crime.

```python
list_crimes = most_common.index
for crime_code in list_crimes:
    y = pd.DataFrame(crimes[crimes['OFFENSE_CODE_GROUP'] == '{}'.format(crime_code)]['STREET'].value_counts().sort_values(ascending=False).head(1))
    print(y)
```

```python
crimes.Lat.replace(to_replace = -1, value = None, inplace=True)    ##Replace -1 values
crimes.Long.replace(to_replace = -1, value = None, inplace=True)
plt.figure(figsize = (10, 10))
sns.scatterplot(x='Lat', 
                y='Long',
                alpha = 0.009,
                data=crimes)
plt.xlim(crimes['Lat'].min(), crimes['Lat'].max())
plt.ylim(crimes['Long'].min(), crimes['Long'].max())
plt.show()
```
